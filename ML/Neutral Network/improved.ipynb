```python
import numpy as np
import matplotlib.pyplot as plt

# Generate large vector data
np.random.seed(42)  # For reproducibility
X = np.random.rand(1_000_000) * 10  # 1 million random data points in the range [0, 10]
y = 2 * X + 3  # Linear relationship with some noise

# Add noise to y
y += np.random.randn(1_000_000) * 0.5
```

```python
# Initialize parameters
epochs = 20
lr = 0.01
w = np.random.rand()  # Random initialization of weight
b = np.random.rand()  # Random initialization of bias
eList = []  # To store error for each epoch

def print_details(epoch, w, b, error):
    """Print training details for each epoch."""
    print(f"Epoch {epoch + 1}:")
    print(f"  Weight: {w:.6f}")
    print(f"  Bias: {b:.6f}")
    print(f"  Error: {error:.6f}\n")
```

```python
# Training loop
for epoch in range(epochs):
    # Forward pass
    y_out = X * w + b
    
    # Compute the loss (mean squared error)
    error = np.mean(np.square(y - y_out))
    eList.append(error)
    
    # Compute the gradients
    gradient_w = -2 * np.mean((y - y_out) * X)
    gradient_b = -2 * np.mean(y - y_out)
    
    # Print details for the current epoch
    print_details(epoch, w, b, error)
    
    # Update parameters
    w -= lr * gradient_w
    b -= lr * gradient_b
```

```python
# Plot epochs vs. error
plt.figure(figsize=(10, 6))
plt.plot(range(1, epochs + 1), eList, marker='o', color='b', label="Error")
plt.title("Epochs vs. Error (Large Dataset)")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.legend()
plt.show()
```
