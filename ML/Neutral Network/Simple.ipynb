```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
# Initialize parameters
epochs = 20
lr = 0.1
w = np.random.rand()  # Random initialization of weight
eList = []  # To store error for each epoch

def print_details(epoch, w, error, gradient):
    """Print training details for each epoch."""
    print(f"Epoch {epoch + 1}:")
    print(f"  Weight: {w:.4f}")
    print(f"  Error: {error:.4f}")
    print(f"  Gradient: {gradient:.4f}\n")
```

```python
# Training loop
for epoch in range(epochs):
    # Forward pass
    y_out = X * w
    
    # Compute the loss (mean squared error)
    error = np.mean(np.square(y - y_out))
    eList.append(error)
    
    # Compute the gradient
    gradient = -2 * np.mean((y - y_out) * X)
    
    # Print details for the current epoch
    print_details(epoch, w, error, gradient)
    
    # Update the weight
    w = w - lr * gradient
```

```python
# Plot epochs vs. error
plt.figure(figsize=(8, 6))
plt.plot(range(1, epochs + 1), eList, marker='o', color='b', label="Error")
plt.title("Epochs vs. Error")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.grid(True)
plt.legend()
plt.show()
```
