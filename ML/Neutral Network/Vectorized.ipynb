```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random multidimensional data
np.random.seed(42)
n_samples = 1_000_000
n_features = 5

X = np.random.rand(n_samples, n_features) * 10  # 1M samples, 5 features
true_weights = np.array([2, -1, 3, 0.5, 4])  # True weights for each feature
true_bias = 3  # True bias

# Generate target values with noise
y = X @ true_weights + true_bias + np.random.randn(n_samples) * 0.5

# Initialize parameters
epochs = 20
lr = 0.01
W = np.random.rand(n_features)  # Random weights
b = np.random.rand()  # Random bias
eList = []  # Store loss for each epoch
```

```python
# Training loop
for epoch in range(epochs):
    # Forward pass: vectorized computation
    y_out = X @ W + b  # Predicted values
    
    # Compute the loss (mean squared error)
    error = np.mean((y - y_out) ** 2)  # Vectorized loss computation
    eList.append(error)
    
    # Compute gradients: vectorized computation
    gradient_W = -2 * (X.T @ (y - y_out)) / n_samples
    gradient_b = -2 * np.mean(y - y_out)
    
    # Update parameters
    W -= lr * gradient_W
    b -= lr * gradient_b
    
    # Print details for the epoch
    print(f"Epoch {epoch + 1}: Loss = {error:.6f}, Weights = {W}, Bias = {b:.6f}")
```

```python
# Plot epochs vs. error
plt.figure(figsize=(10, 6))
plt.plot(range(1, epochs + 1), eList, marker='o', color='b', label="Error")
plt.title("Epochs vs. Error (Multidimensional Data)")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.legend()
plt.show()
```
